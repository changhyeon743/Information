```python
import csv
f = open('2008년 연령별 인구.csv', encoding = 'cp949')
data = csv.reader(f)

result1 = []

for row in data:
    if '서울특별시  (1100000000)' in row[0]:
        for i in range(3, len(row)):
            temp = ''.join(row[i].split(','))
            result1.append(int(temp))
            
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.plot(result1)
plt.title('2008년 연령별 인구')
plt.show()
```

```python
import csv
f = open('2019년 연령별 인구.csv', encoding = 'cp949')
data = csv.reader(f)

result2 = []

for row in data:
    if '서울특별시  (1100000000)' in row[0]:
        for i in range(3, len(row)):
            temp = ''.join(row[i].split(','))
            result2.append(int(temp))
            
import matplotlib.pyplot as plt
plt.plot(result2)
plt.title('2019년 연령별 인구')
plt.show()
```
