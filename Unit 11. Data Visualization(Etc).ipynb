## 2. 막대그래프 (bar chart)

### 1) 막대 그래프 : bar()

```python
import matplotlib.pyplot as plt

plt.bar([1, 2, 3, 4], [5, 67, 100, 30])
plt.show()
```

### 2) 막대의 폭 지정 : width

```python
burger = [6, 10, 8]

plt.bar(range(3), burger, width = 0.4)
plt.show()
```

### 3) 틱 설정 : 축 상의 위치 표시 지점, xticks(), yticks()

```python
brand = ['B-Burger', 'L-Burger', 'M-Burger']
burger = [6, 10, 8]

plt.bar(range(3), burger, width = 0.4)
plt.xticks(range(3), brand)
plt.yticks([6, 8, 10])
plt.show()
```

### 4) 그리드 설정 : grid()

```python
brand = ['B-Burger', 'L-Burger', 'M-Burger']
burger = [6, 10, 8]

plt.bar(range(3), burger, width = 0.4)
plt.xticks(range(3), brand)
plt.yticks([6, 8, 10])
plt.grid(True)
plt.show()
```

### 5) 그림 범위 지정 : xlim, ylim

```python
import matplotlib.pyplot as plt

plt.bar([1, 2, 3, 4], [10, 20, 30, 40])
plt.xlim(0, 5)
plt.ylim(-2, 50)
plt.show()
```

### 6) 범례 설정

```python
brand = ['B-Burger', 'L-Burger', 'M-Burger']
burger = [6, 10, 8]
pizza = [14, 20, 19]

plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(3), burger, width = 0.4, label = '햄버거가격')
plt.bar([0.5, 1.5, 2.5], pizza, width = 0.4, label = '피자가격')
plt.legend()
plt.xticks([0.25, 1.25, 2.25], brand)
plt.show()
```

### 7) 수평 막대 그래프 : barh()

```python
brand = ['B-Burger', 'L-Burger', 'M-Burger']
burger = [6, 10, 8]
plt.barh(range(3), burger, height = 0.4)
plt.yticks(range(3), brand)
plt.show()
```

## 3. 히스토그램 (histogram)

- 도수분포 : 표본의 다양한 산출 분포를 나타낸 것. 각 항목은 특정 그룹이나 주기 안에 값이 발생한 빈도나 횟수를 포함한다.
- 히스토그램 : 도수분포표를 그래프로 나타낸 것
- hist() : 히스토그램 그리기, 데이터 집계 결과를 반환한다.
- bins 인수로 데이터를 집계할 구간 정보를 받는다.

```python
import matplotlib.pyplot as plt

plt.hist([1, 1, 2, 4, 5, 6, 7, 8, 9], bins = 9)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist([1, 1, 2, 4, 5, 7, 8, 9], bins = 3)
plt.xticks([2, 5, 8], ['1~3','4~6','7~9'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

result = []
for i in range(1000000):
    result.append(random.randint(1,6))
    
plt.rc('font', family='Malgun Gothic')
plt.hist(result, bins=6, color='g')
plt.title('주사위 시뮬레이션 결과(1,000,000번)')
plt.show()
```

## 4. 상자 수염 그림 (boxplot)

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
z = []

for i in range(100):
    x.append(random.randint(1,100))
    y.append(random.randint(1,100))
    z.append(random.randint(1,100))
    
plt.figure(dpi=150)
plt.title('평가 결과 분석')
plt.boxplot([x, y, z], labels = ['1반', '2반', '3반'])
plt.show()
```

## 5. 산포도 (scatter)

- 산포도 : 자료의 분산 상황을 나타내는 수의 값. 변량이 분포의 중심값에 흩어진 정도.

```python
import matplotlib.pyplot as plt


plt.scatter([1, 2, 3, 2], [2, 3, 3, 1])
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []

for i in range(100):
    x.append(random.randint(1, 10000))
    y.append(random.randint(1, 10000))
    
plt.title('랜덤 좌표')
plt.scatter(x, y)
plt.show()
```

 점의 크기 모양 변경
 - s='숫자' : 점의 크기 
 - c='색상' : 점의 색깔 
 - edgecolor='색상 : 점의 테두리 색상 
 - alpha='숫자' : 점의 투명도

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
z = []

for i in range(100):
    x.append(random.randint(1, 10000))
    y.append(random.randint(1, 10000))
    z.append(random.randint(1, 1000))
    
plt.scatter(x, y, z, c = ['r', 'y', 'm', 'c']*25, alpha = 0.7)
plt.title('랜덤 좌표, 랜덤 사이즈')
plt.show()
```

## 6. 원 그래프 (pie)

```python
import matplotlib.pyplot as plt

size = [15, 20, 25, 40]
plt.pie(size)
plt.axis('equal')
plt.show()
```

파이차트 그리기
- size : 파이 차트를 구성할 데이터
- explode : 파이 조각이 돌출되는 크기. 0이면 돌출되지 않음
- labels : 파이 조각의 라벨
- colors : 파이 색깔
- autopct : 파이 조각의 전체 대비 백분율
- startangle : 그려지는 파이 조각 시작 위치. 90이면 12시 방향임. 파이 조각은 이 각도 기준을 반시계 방향으로 그려짐
- shadow : 파이 차트의 그림자 효과 유무

```python
import matplotlib.pyplot as plt

size = [231, 132, 94, 341]
explode = (0, 0, 0, 0.2)
labels = ['기린', '고양이', '토끼', '강아지']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
plt.figure(figsize=(6,4), dpi=300)
plt.pie(size, explode = explode, labels = labels, colors = colors, startangle = 90, autopct = '%1.1f%%')
plt.title('좋아하는 동물 투표 결과')
plt.show()
```
